```python
from selenium import webdriver #pip install selenium if you have not already
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time
```

```python
path = r"C:\Users\sanji\Downloads\chromedriver_win32 (5)\chromedriver.exe" # check hrome browser version and then install the required chrome driver and write down the path here
driver = webdriver.Chrome(path) 

input_suburb_name = ['greendale','gregory hills', 'leppington', 'oran park', 'denham court']


def get_Data(data):

    driver.get('https://dbr.abs.gov.au/index.html')
    search = driver.find_element_by_class_name('searchInput')
    search.send_keys(f'{data}')
    search.send_keys(Keys.ENTER)
    time.sleep(1)
    website_click = driver.find_element_by_xpath('//*[@id="pnlSearch"]/div/div[2]').click()
    time.sleep(1)


    web_click = driver.find_element_by_xpath('//*[@id="mapDiv_root"]/div[3]/div[1]/div[2]/div/table/tbody/tr[5]/td[4]/a').click()

    time.sleep(1)



    x = driver.find_element_by_class_name('biscuitTable')
    #getting erp row
    erp = x.find_element_by_id('ERP_P_20name').text
    erp_year = x.find_element_by_id('ERP_P_20year').text
    erp_region = x.find_element_by_id('ERP_P_20data').text


    #getting median total income excluding government pension and allowance
    mti = x.find_element_by_id('INCOME_17name').text
    mti_year = x.find_element_by_id('INCOME_17year').text
    mti_region = x.find_element_by_id('INCOME_17data').text



    #median weekly household rent
    mwhp = x.find_element_by_id('RENT_4name').text
    mwhp_year = x.find_element_by_id('RENT_4year').text
    mwhp_region = x.find_element_by_id('RENT_4data').text


    #Median monthly household mortgage payment ($)
    mmhmp = x.find_element_by_id('RENT_5name').text
    mmhmp_year = x.find_element_by_id('RENT_5year').text
    mmhmp_region = x.find_element_by_id('RENT_5data').text


    draft_table_dict = { 'suburb_name' : data, 
                        'description' : [mti, mwhp, mmhmp], 
                        'year': [mti_year, mwhp_year,mmhmp_year], 
                        'data': [ mti_region, mwhp_region, mmhmp_region]}
    df = pd.DataFrame(draft_table_dict)
    return df
    



suburb = []
for i in input_suburb_name:
    suburb.append(get_Data(i))

driver.quit() 
 
df = pd.concat(suburb) 
df.reset_index(drop=True, inplace=True)
df.index += 1
print(df)


```
